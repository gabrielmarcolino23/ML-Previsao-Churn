```python
account.head()
```

```python
dados = pd.concat([df['customerID'], df['Churn'],customer, phone, internet, account], axis=1) # concatenando todas as colunas
```

```python
dados.head()
```

```python
# renomeando as colunas
dados.columns = ['ID_Cliente', 'Churn', 'Genero', 'Idoso', 'Parceiro', 'Dependentes', 'Meses_na_Empresa', 'Servico_Telefonico', 'Multiplas_Linhas', 'Servico_Internet', 'Servico_Seguranca_Online', 'Backup_Online', 'Protecao_Dispositivo', 'Suporte_Tecnico', 'TV_Streaming', 'Filmes_Streaming', 'Contrato', 'Faturamento_Eletronico', 'Metodo_Pagamento', 'Cobranca_Mensal', 'Cobranca_Total']
```

```python
dados.head()
```

```python
dados.info()
```

```python
# Convertendo coluna Idoso para objeto e transformando os valores 0 e 1 em No e Yes
dados['Idoso'] = dados['Idoso'].astype('object')
dados['Idoso'] = dados['Idoso'].map({0: 'No', 1: 'Yes'})
```

```python
dados.head()
```

```python
dados['Idoso'].unique()
```

```python
dados['Cobranca_Total'].head(500)
```

```python
dados['Cobranca_Total'] = dados['Cobranca_Total'].str.strip() # removendo espaços em branco
```

```python
#print das linhas onde a coluna Cobranca_Total é vazia
dados[dados['Cobranca_Total'] == '']
```

```python
# Substituir strings vazias por 0
dados['Cobranca_Total'] = dados['Cobranca_Total'].replace('', '0')
```

```python
# Contandos os nan
dados['Cobranca_Total'].isnull().sum()
```

```python
dados['Cobranca_Total'] = dados['Cobranca_Total'].astype(float) 
```

```python
dados.info()
```

```python
dados['ID_Cliente'].duplicated().sum()
```

```python
for coluna in dados.columns:
    print('---')
    print(dados[coluna].value_counts())
```

```python
# Removendo valores vazios em Churn
dados = dados[dados['Churn']!= '']
dados.reset_index(drop=True, inplace=True)
```

```python
# Criação de coluna contendo os gastos diarios do cliente na posição 18
dados['Cobranca_Diaria'] = dados['Cobranca_Total'] / dados['Meses_na_Empresa']
```

```python
cols = dados.columns.to_list()
cols.remove('Cobranca_Diaria')
```

```python
cols.insert(19, 'Cobranca_Diaria')
```

```python
dados = dados[cols]
```

```python
dados['Cobranca_Diaria'] = dados['Cobranca_Diaria'].round(2)
```

```python
dados.to_json("Telco-Customer-Churn-limpeza.json")  
```
